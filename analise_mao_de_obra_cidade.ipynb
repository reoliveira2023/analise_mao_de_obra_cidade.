```python
import pandas as pd
import numpy as np

# Definir quantidade de habitantes
populacao_total = 50000

# Criar IDs únicos para cada habitante
ids = np.arange(1, populacao_total + 1)

# Sexo: 50% masculino, 50% feminino
sexo = np.random.choice(['Masculino', 'Feminino'], size=populacao_total, p=[0.5, 0.5])

# Faixa etária:
# Para simplificar, vamos gerar:
# - Masculinos → 20% na faixa de 18 a 60 com capacidade para trabalho operacional
# - Femininos → não entram na análise de trabalho operacional neste cenário

# Inicializar idade com valores aleatórios
idade = np.random.randint(0, 90, populacao_total)

# Definir capacidade de trabalho
capacidade_trabalho = []
for s, i in zip(sexo, idade):
    if s == 'Masculino' and (18 <= i <= 60):
        capacidade_trabalho.append('Apto')
    else:
        capacidade_trabalho.append('Não Apto')

# Status de ocupação (apenas para aptos masculinos):
# 8% dos aptos estão desempregados
status_ocupacao = []
for cap in capacidade_trabalho:
    if cap == 'Apto':
        status_ocupacao.append(
            np.random.choice(['Desempregado', 'Empregado'], p=[0.08, 0.92])
        )
    else:
        status_ocupacao.append('Não se aplica')

# Alfabetização (apenas para desempregados aptos):
# Apenas 5% sabem ler e escrever
alfabetizado = []
for cap, ocup in zip(capacidade_trabalho, status_ocupacao):
    if cap == 'Apto' and ocup == 'Desempregado':
        alfabetizado.append(
            np.random.choice(['Sim', 'Não'], p=[0.05, 0.95])
        )
    else:
        alfabetizado.append('Não se aplica')

# Criar DataFrame
df = pd.DataFrame({
    'ID': ids,
    'Sexo': sexo,
    'Idade': idade,
    'Capacidade_Trabalho': capacidade_trabalho,
    'Status_Ocupacao': status_ocupacao,
    'Alfabetizado': alfabetizado
})

# Visualizar amostra dos dados
df.head(50)
```
